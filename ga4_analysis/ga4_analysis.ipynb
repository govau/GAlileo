```python
# load the data
data = []
for line in open('nikis_data_analysis_ga4_events_dta', 'r'):
    data.append(json.loads(line))
```

```python
# events per user 
# events per day
# engaged sessions/ session per day
# 
```

```python
total_events = 0
total_users = 0
total_sessions = 0
total_engaged_sessions = 0
sessions_per_day = []
engaged_sessions_per_day = []
unengaged_sessions_per_day = []
users_per_day = []

for row in range(len(df)):
    current_date = df['event_date'][row]
    current_user = df['user_pseudo_id'][row]
    current_event = df['event_name'][row]
    event_param_key = df['event_param_key'][row]
    event_param_value = df['event_param_value'][row]
    
    if row == 0:
        total_events = 1
        total_users = 1
        sessions_today = 1
        total_sessions = 1
        if df['event_param_key'][row]=='sessions_engaged' and df['event_param_value'][row] == 1:
            total_engaged_sessions = 1
            engaged_sessions_today = 1
        else:
            total_engaged_sessions = 0
            unengaged_sessions_today = 1
            engaged_sessions_today = 0
    elif row > 0:
        previous_row =row-1
        # sessions per day/ users per day
        if current_date == df['event_date'][previous_row] and current_user != df['user_pseudo_id'][previous_row]:
            total_sessions += 1
            sessions_today += 1
        elif current_date != df['event_date'][previous_row]:
            sessions_per_day.append(sessions_today)
            sessions_today = 0
        # engaged sessions per day  
        if current_date == df['event_date'][previous_row] and event_param_key == 'session_engaged':
            if event_param_value == 1:
                engaged_sessions_today += 1
                total_engaged_sessions +=1
            elif event_param_value == 0:
                unengaged_sessions_today +=1
        elif current_date != df['event_date'][previous_row]:
            engaged_sessions_per_day.append(engaged_sessions_today)
            engaged_sessions_today = 0
            unengaged_sessions_per_day.append(unengaged_sessions_today)
            unengaged_sessions_today = 0
            
```
